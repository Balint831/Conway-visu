```python
import pandas as pd
import matplotlib.pyplot as plt

mp = pd.read_csv("times_mp.txt",header = None)
mp = mp.T

st = pd.read_csv("times.txt", header = None)
st = st.T

gpu = pd.read_csv("gpu_runtimes.txt", header=None)
gpu = gpu.T # it is in milliseconds
```

```python
st.columns = st.iloc[0]
st.drop(index = [0,11], inplace = True)
```

```python
gpu.columns = gpu.iloc[0]
gpu.drop(index = [0,11], inplace = True)
```

```python
mp.columns = mp.iloc[0]
mp.drop(index = [0,11], inplace = True)
```

```python
plt.figure(figsize=[16,6])
plt.subplot(1,2,1)
plt.plot(st.min(axis=0), color="tab:red", label="single thread")
plt.plot(mp.min(axis=0), color ="tab:orange", label="multi thread")

plt.plot(gpu.min(axis=0)*1000, color="tab:purple", label="GPU")
plt.yscale("log")
plt.xscale("log")
plt.title("CPU multithreading", size =16)
plt.ylabel(r"min runtime [$\mu s$]", size=16)
plt.xlabel(r"$\sqrt{grid size}$", size=16)
plt.grid()
plt.legend()

plt.subplot(1,2,2)
plt.plot(st.min(axis=0), color="tab:red", label="single thread")
plt.plot(mp.min(axis=0), color ="tab:orange", label="multi thread")

plt.plot(gpu.min(axis=0)*1000, color="tab:purple", label="GPU")

plt.title("CPU multithreading", size =16)
plt.ylabel(r"min runtime [$\mu s$]", size=16)
plt.xlabel(r"$\sqrt{grid size}$", size=16)
plt.grid()
plt.legend()
plt.savefig("pics/comparison.png")
```
